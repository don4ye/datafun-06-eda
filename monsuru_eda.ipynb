# Initial Data Distribution for Numerical Columns
In this section, we will visualize the data distribution for numerical columns in the dataset using histograms.

```python
df.hist()
plt.show()
```

## Observations from Histograms

- **Carat Weight**: The distribution of carat weight appears to be right-skewed, with a majority of diamonds having lower carat weights and fewer diamonds with higher carat weights.
- **Price**: The distribution of price also appears to be right-skewed, indicating that most diamonds have lower prices, with a few diamonds having significantly higher prices.
- **Depth and Table**: The distributions of depth and table seem to be relatively symmetric, with no apparent skewness.
- **Dimensions**: Histograms for dimensions (x, y, z) show variations in distribution, which may reflect differences in the size and shape of diamonds.
- **Observations on Outliers**: There are potential outliers in the histograms of carat weight and price, as indicated by the presence of data points far from the main bulk of the distribution. These outliers may represent exceptional cases or data errors and may need further investigation.
- **Data Quality**: Overall, the histograms provide a good overview of the distribution of numerical features in the diamonds dataset. The data appears to be relatively clean, with no obvious anomalies in the distributions.


# Initial Data Distribution for Categorical Columns
In this section, we will explore the distribution of data across categorical columns in the dataset.

### Value Counts for Individual Categorical Column (Species)

Let's start by inspecting the value counts for the 'species' column.

```python
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()
```

## Observations from Value Counts and Count Plots

### Species Distribution
- The 'species' column represents the different species of diamonds. The value counts reveal the frequency of each species in the dataset.

### Distribution of Other Categorical Columns
- **Cut**: The distribution of diamond cuts shows that 'Ideal' and 'Premium' cuts are more common compared to other categories.
- **Color**: The distribution of diamond colors indicates that 'G' and 'E' colors are more prevalent, while 'J' color is less common.
- **Clarity**: The distribution of diamond clarity reveals that 'SI1' and 'VS2' clarities are the most frequent categories.
- **Observations on Imbalance**: There seems to be an imbalance in some categorical features. For example, in the 'color' column, certain colors are more dominant than others, which might affect the analysis if not addressed properly.
